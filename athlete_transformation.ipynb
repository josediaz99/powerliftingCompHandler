### transforming using pickle transformation

```python
import csv
import pickle

csv_file = r'data\open_nationals_day_2_awards_results.csv'
pickle_file = r'data\open_nationals_day_2_awards_results.pkl'


try:
    with open(csv_file, 'r', encoding='utf-8') as f:
        reader = csv.reader(f)
        data_as_list = list(reader) # Read all data into a list

    with open(pickle_file, 'wb') as f: # Open in write binary mode
        pickle.dump(data_as_list, f)
    
    print(f"Successfully converted {csv_file} to a list and saved as {pickle_file} using built-in modules.")

except FileNotFoundError:
    print(f"Error: {csv_file} not found.")
except Exception as e:
    print(f"An error occurred: {e}")
```
